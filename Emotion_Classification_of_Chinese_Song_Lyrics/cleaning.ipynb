```python
import pandas as pd
import re
```

```python
data = pd.read_csv("data/final.csv")
data.head()
```

```python
clean_text = []
for i in data["text"]:
    #清除metadata
    step1 = re.sub(r"(原*作詞|改編詞|OT|原*作曲|原*編曲|製作人*|監製|唱|[大小]提琴|和聲|[錄混]音|吉他|弦樂(編寫|監製)|企劃|母帶)：(\S+ *\S+|(\S+ \/ \S+)+), ", " ", i)
    #清除featuring
    step2 = re.sub(r"feat. \S+\s*\S+, ", "", step1)
    #清除戲劇資訊
    step3 = re.sub(r"^.+((主題|片[頭尾]|插)曲|配樂), ", " ", step2)
    #清除資料提供者
    step4 = re.sub(r", 感謝, \S+, (提供|修正)歌詞", "", step3)
    #清除其他metadata
    step5 = re.sub(r"(ti|ar|al|by|offset):(\S+ *\S+|(\S+ \/ \S+)+), ", "", step4)
    #清除分part
    step6 = re.sub(r"\S+ *\S+[:：], ", "", step5)
    #清除亂碼
    step7 = re.sub(r"[\u200b|\u202c|\u202d]+", "", step6)
    #清除非中英文、逗點、空格字元
    step8 = re.sub(r"[^\u4e00-\u9fff\s,A-z]+", "", step7)
    #清除重複空格
    step9 = re.sub(r"\s{2,}", " ", step8)
    #清除最開頭空格及逗點
    step10 = re.sub(r"^[\s,]+", "", step9)
    clean_text.append(step10)
```
